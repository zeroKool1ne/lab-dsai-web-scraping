#### 2. Display all the image links from Walt Disney wikipedia page.
Hint: use `.get()` to access information inside tags. Check out the documentation.

````python
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

images = soup.find_all('img')

image_links = []
for img in images:
    src = img.get('src')
    if src:
        if src.startswith('//'):
            src = 'https:' + src
        image_links.append(src)

for link in image_links:
    print(link)
````

#### 2.1. List all language names and number of related articles in the order they appear in wikipedia.org.

````python
url = 'https://www.wikipedia.org/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

lang_elements = soup.find_all('a', attrs={'class': 'link-box'})

languages = []
for lang in lang_elements:
    name = lang.find('strong')
    count = lang.find('small')
    if name and count:
        languages.append((name.get_text(strip=True), count.get_text(strip=True)))

for name, count in languages:
    print(f"{name}: {count}")
````

#### 3. Display Metacritic top 24 Best TV Shows of all time (TV Show name, initial release date, metascore rating, film rating system and description) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url = 'https://www.metacritic.com/browse/tv/'
````

````python
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
print(f"Status code: {response.status_code}")

soup = BeautifulSoup(response.text, 'html.parser')

titles = []
dates = []
metascores = []
descriptions = []

cards = soup.find_all('div', attrs={'class': 'c-finderProductCard'})

for card in cards[:24]:
    title_tag = card.find('h3', attrs={'class': 'c-finderProductCard_titleHeading'})
    titles.append(title_tag.get_text(strip=True) if title_tag else None)

    date_tag = card.find('span', attrs={'class': 'u-text-uppercase'})
    dates.append(date_tag.get_text(strip=True) if date_tag else None)

    score_tag = card.find('div', attrs={'class': 'c-siteReviewScore'})
    metascores.append(score_tag.get_text(strip=True) if score_tag else None)

    desc_tag = card.find('div', attrs={'class': 'c-finderProductCard_description'})
    descriptions.append(desc_tag.get_text(strip=True) if desc_tag else None)

df = pd.DataFrame({
    'Title': titles,
    'Release Date': dates,
    'Metascore': metascores,
    'Description': descriptions
})

df
````

#### 3.1. Find the image source link and the TV show link. After you're able to retrieve, add them to your initial dataframe

````python
image_links = []
show_links = []

for card in cards[:24]:
    img_tag = card.find('img')
    image_links.append(img_tag.get('src') if img_tag else None)

    link_tag = card.find('a')
    href = link_tag.get('href') if link_tag else None
    if href and not href.startswith('http'):
        href = 'https://www.metacritic.com' + href
    show_links.append(href)

df['Image Link'] = image_links
df['Show Link'] = show_links

df
````
